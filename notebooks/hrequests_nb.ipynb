```python
print("\n--- HEAD Request ---")
url = 'https://www.google.com/' # Using Google for a real-world example

resp = hrequests.head(url)

print(f"Status Code: {resp.status_code}")
print("Response Headers:")
for key, value in resp.headers.items():
    print(f"  {key}: {value}")

print(f"Response Body (should be empty for HEAD): {resp.text[:100] if resp.text else 'Empty'}") #since resp contains only headers
```

```python
'''OPTIONS requests are used to describe the communication options for the target resource. 
It allows a client to discover the methods and other options supported by a web server for a particular URL without actually performing an action.'''

print("\n--- OPTIONS Request ---")
url = 'https://httpbin.org/anything' # httpbin.org/anything supports OPTIONS

resp = hrequests.options(url)

print(f"Status Code: {resp.status_code}")
print("Response Headers (focus on 'Allow' header):")
for key, value in resp.headers.items():
    print(f"  {key}: {value}")

print(f"Response Body (often empty or minimal for OPTIONS): {resp.text[:100] if resp.text else 'Empty'}")
```

```python
import hrequests
import json # Good practice to import json for potential pretty printing or handling

# 1. Make a request to an endpoint that returns JSON
# For demonstration, httpbin.org/json returns a simple JSON object
url_json_response = 'https://httpbin.org/json'
resp = hrequests.get(url_json_response)

print(f"Status Code: {resp.status_code}")
print(f"Content-Type: {resp.headers.get('Content-Type')}")

# 2. Use the .json() method to parse the response body
try:
    json_data = resp.json()

    print("\nParsed JSON Data:")
    # Pretty print the JSON for better readability
    print(json.dumps(json_data, indent=2))

    # You can now access elements of the JSON data like a Python dictionary or list
    if isinstance(json_data, dict) and 'slideshow' in json_data:
        print(f"\nTitle from JSON: {json_data['slideshow']['title']}")

except json.JSONDecodeError:
    print(f"Error: Could not decode JSON from response. Response text:\n{resp.text[:500]}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")

# Example 2: What if the response is not JSON?
print("\n--- Attempting to parse non-JSON response ---")
url_html_response = 'https://www.google.com/'
resp_html = hrequests.get(url_html_response)

print(f"Status Code: {resp_html.status_code}")
print(f"Content-Type: {resp_html.headers.get('Content-Type')}")

try:
    # This will likely raise a JSONDecodeError
    json_data_html = resp_html.json()
    print("Parsed HTML as JSON:", json_data_html) # This line might not be reached
except json.JSONDecodeError:
    print("Successfully caught JSONDecodeError: Response was not valid JSON.")
    print(f"First 500 characters of HTML response:\n{resp_html.text[:500]}...")

# By using json_data = resp.json() (and adding error handling), 
# we will correctly get our response body parsed into a Python object, which we can then work with.
```

```python

print("\n--- GET with 'cookies' (Corrected approach using Session) ---")

# Create a session object
s = hrequests.Session()

url = 'https://httpbin.org/cookies'
my_cookies = {'mycookie': 'myvalue', 'session_id': 'abc123'}

# Add cookies to the session's cookie jar
# This is the correct way to seed a session with initial cookies
s.cookies.update(my_cookies)

# Now, make the request using the session. The session will automatically
# send the cookies in its jar.
resp = s.get(url)

print(f"Status Code: {resp.status_code}")
print("Response JSON:")
print(json.dumps(resp.json(), indent=2))
```

# Sessions

```python
# Create a session object mimicking a random browser version
s = hrequests.Session()
print(f"Session User-Agent (randomized): {s.headers['User-Agent']}")

# First request - server might set cookies
print("\n--- First Request (to set a cookie) ---")
# httpbin.org/cookies/set sets a cookie 'name' with value 'value'
resp1 = s.get('https://httpbin.org/cookies/set?name=myvalue')
print(f"Status Code: {resp1.status_code}")
print(json.dumps(resp1.json(), indent=2))
print(f"Cookies from resp1: {s.cookies.get_dict()}") # s.cookies stores the received cookie
```

```python
print(f"Cookies currently in session 's' after resp1: {s.cookies.get_dict()}")
```

### The key here is that cookies are domain and path specific. 
The cookie name=myvalue set by httpbin.org is only valid for the httpbin.org domain (and potentially specific paths on that domain). When you call s.cookies.get_dict() without specifying a URL or context, it might only return cookies that are truly "global" or those that are currently active for a hypothetical default context, it's designed to return the simplest representation of cookies it can, often excluding those that are bound to specific hosts or paths if the current context doesn't match.

```python
# Second request - the session should automatically send the cookie back
print("\n--- Second Request (to verify cookie is sent by session) ---")
# httpbin.org/cookies echoes back received cookies
resp2 = s.get('https://httpbin.org/cookies')

print(f"Status Code: {resp2.status_code}")
print("Response JSON from httpbin.org/cookies (showing received cookies):")
print(json.dumps(resp2.json(), indent=2))

# Verify that the session's cookie jar now contains the cookie (after it has been used)
print(f"Cookies in session 's' after resp2: {s.cookies.get_dict()}")

# You can also set default headers for the session
s.headers.update({'X-My-Custom-Header': 'HrequestsSessionTest'})
print(f"\nSession Headers after update: {s.headers}")
```

```python
print("\n--- Third Request (with custom session header) ---")
resp3 = s.get('https://httpbin.org/headers')
print(f"Status Code: {resp3.status_code}")
print("Response JSON for headers:")
print(json.dumps(resp3.json(), indent=2))
```

```python
# Example using raw Playwright in Jupyter
import asyncio
from playwright.async_api import async_playwright

async def run_playwright():
    async with async_playwright() as p:
        browser = await p.firefox.launch() # or .chromium.launch()
        page = await browser.new_page()
        await page.goto("https://www.google.com")
        print(await page.title())
        await browser.close()

await run_playwright()
```

```python
# Now you can proceed with your hrequests browser automation code
# For example:
import asyncio
import time

async def automate_browser_task():
    page = None
    try:
        session = hrequests.Session()
        print("hrequests Session created.")
        page = await session.render('https://www.google.com/')
        print(f"Browser page rendered for: {page.url}")
        await asyncio.sleep(2) # Use asyncio.sleep for async delays
        print("Finished waiting.")
    except Exception as e:
        print(f"An error occurred: {e}")
    finally:
        if page and not page.is_closed():
            print("Closing browser page...")
            page.close()
            print("Browser page closed.")
        else:
            print("Browser page was not opened or already closed.")

# Execute the async function
await automate_browser_task()

print("\nAutomation task finished.")
```

```python
import nest_asyncio
import hrequests
import asyncio
import time

# Apply the nest_asyncio patch: This *must* be the first async-related thing you do in a Jupyter session.
# It allows hrequests's internal asyncio.run() calls to work within Jupyter's existing event loop.
print("Applying nest_asyncio patch...")
nest_asyncio.apply()
print("nest_asyncio patch applied.")

async def test_hrequests_browser_automation():
    """
    This is our main asynchronous function for browser automation.
    It simulates a simple browser interaction.
    """
    page = None # Initialize page variable for clean error handling
    try:
        # Create an hrequests Session.
        # This part doesn't directly involve browser launching yet.
        session = hrequests.Session()
        print("hrequests Session created.")

        # This is where the browser is launched and the URL is visited.
        # 'await' is used because session.render() is an asynchronous operation.
        # Internally, hrequests's session.render() method will make calls that need the event loop.
        print("Attempting to render Google.com...")
        page = await session.render('https://www.scrapingcourse.com/infinite-scrolling')
        print(f"Browser page rendered for: {page.url}")

        # Simulate waiting for the page to load or for user interaction.
        # asyncio.sleep is used for non-blocking delays in async code.
        print("Waiting for 2 seconds...")
        await asyncio.sleep(2)
        print("Finished waiting.")

        # Example interaction: Print the page title
        title_element = page.find('title')
        if title_element:
            print(f"Page Title: {title_element.text}")
        else:
            print("Could not retrieve page title.")

    except Exception as e:
        # Catch any errors that occur during the automation
        print(f"An error occurred: {e}")
    finally:
        # Ensure the browser page is closed regardless of success or failure
        if page and not page.is_closed():
            print("Closing browser page...")
            page.close()
            print("Browser page closed successfully.")
        else:
            print("Browser page was not opened or already closed, no action needed.")

print("\nStarting the browser automation task...")
# Execute the asynchronous function.
# In Jupyter, you can directly 'await' a top-level async function.
# This tells Jupyter's existing asyncio event loop to run 'test_hrequests_browser_automation'.
await test_hrequests_browser_automation()

print("\nAutomation task completed.")
```
